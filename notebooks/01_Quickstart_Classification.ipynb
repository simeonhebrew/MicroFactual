# 📘 01. Quickstart Classification

Welcome to **MicroFactual**! This notebook demonstrates how to run a complete microbiome classification pipeline in less than 5 minutes using the high-level API.

**Goal**: Classify Colorectal Cancer (CRC) samples from healthy controls.
**Dataset**: Public CRC dataset (included in `datasets/`).

```python
import microfactual as mf
import pandas as pd
import matplotlib.pyplot as plt

# Enable matplotlib inline plotting
%matplotlib inline
```

## 1. the One-Liner API

The `mf.classify()` function handles everything:
1. Loading data
2. Filtering low-abundance features
3. CLR transformation
4. Training (Random Forest by default)
5. Cross-validation

```python
# file paths
abundance_file = "../datasets/abundance_crc.txt"
metadata_file = "../datasets/metadata_crc.txt"

# Run classification
results = mf.classify(
    abundance_file,
    metadata_file,
    target_column="Group",
    sample_column="Sample ID",
    algorithm="random_forest",
    cv_folds=5
)
```

## 2. Inspect Results

The function returns a dictionary containing the trained model, the processed dataset, and key metrics.

```python
# Check Mean CV scores
print("Cross-Validation Performance:")
for metric, score in results["cv_scores"].items():
    print(f"{metric}: {score:.3f}")
```

## 3. Visualize Performance

We can easily plot the ROC curve for the trained model on the full dataset.

```python
# Get model probabilities on the training data (for demonstration)
model = results["model"]
dataset = results["dataset"]

y_true = dataset.y
y_probs = model.predict_proba(dataset.X)[:, 1]

# Plot ROC
fig = mf.plot_roc(y_true, y_probs, title="ROC Curve (Full Dataset)")
plt.show()
```

```python
# Extract feature names after preprocessing
feature_names = model.pipeline_[:-1].get_feature_names_out(dataset.feature_names)

# Plot Feature Importance
fig = mf.plot_feature_importance(model, feature_names=feature_names, top_n=15)
plt.show()
```
